### 3锔 Calidad de los datos

````python
def calidad (df, nombre, clave=None):
  print(f'calidad de datos de {nombre}:')
  display(df.isna().sum().to_frame('nulos'))
  dup_rows= df.duplicated(keep=False).sum()
  print('las filas duplicadas son:', dup_rows)
  if clave and clave in df.columns:
    dup_key = df[clave].duplicated(keep=False).sum()
    print(f"Duplicados por clave '{clave}':", dup_key)
    if dup_key > 0:
      duplicados_ordenados= (
          df[df[clave].duplicated(keep=False)][clave]
          .value_counts().sort_values(ascending=False)
      )
      print('Duplicados m谩s frecuentes:')
      display(duplicados_ordenados.head(10))
    else:
      print(f' no se encuentraron duplicados con la clas {clave}')
  else:
    if clave:
      print(f'la clave {clave} no existe en el dataframe')
    else:
      print('no se pas贸 una clave')

````

````python
calidad(ventas, 'Calidad de los datos de ventas', clave='id_venta')
calidad(clientes, 'Calidad de los datos de clientes', clave='id_cliente')
calidad(marketing, 'Calidad de los datos de marketing', clave='id_campanha')
````

### 4锔 Limpieza de datos

````python
#creo copias de los dataframe y sobreescribo el objeto eliminando los duplicados
ventas_clean =ventas.copy()
clientes_clean= clientes.copy()
marketing_clean=marketing.copy()

ventas_clean= ventas_clean.drop_duplicates()
clientes_clean = clientes_clean.drop_duplicates()
marketing_clean= marketing_clean.drop_duplicates()

#ahora chequeo la calidad de datos sin duplicados

calidad(ventas_clean, 'Ventas Limpia', clave= 'id_venta')
calidad(clientes_clean, 'Clientes Limpia', clave= 'id_cliente')
calidad(marketing_clean, 'Marketing Limpia', clave= 'id_campanha')
````

````python
#normalizamos el texto
def normalizar_texto(df):
  for col in df.select_dtypes(include='object').columns:
    df[col]= (
        df[col]
        .astype(str)
        .str.replace(r'[\u200b\t\r\n]', "", regex=True)
        .str.replace(' +', " ", regex=True)
        .str.title()
    )
  return df


ventas_clean = normalizar_texto(ventas_clean)
clientes_clean = normalizar_texto(clientes_clean)
marketing_clean = normalizar_texto(marketing_clean)

````

````python
#Normalizar fechas
for df in [ventas_clean, clientes_clean, marketing_clean]:
  for col in df.columns:
    if 'fecha' in col.lower():
      df[col]= pd.to_datetime (df[col], errors='coerce', dayfirst=True)


#imprimo para ver si quedo todo bien

print(ventas_clean.head(10))
print(clientes_clean.head(10))
print(marketing_clean.head(10))
````

````python
#Ahora se normalizan los numericos:

if "precio" in ventas_clean.columns:
    ventas_clean["precio"] = (
        ventas_clean["precio"]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    ventas_clean["precio"] = pd.to_numeric(ventas_clean["precio"], errors="coerce")
print(ventas_clean.dtypes)

#como cantidad qued贸 como float lo paso a int

if "cantidad" in ventas_clean.columns:
  ventas_clean["cantidad"] = pd.to_numeric(
  ventas_clean["cantidad"], errors="coerce"
  ).astype("Int64")


````

````python
#Ahora se crea una funcion para un reporte global donde se seleccoinan todos los data frame (con zip), y por cada uno se van calculando
#todos los nulos, duplicados, columnas, y filas
#la funcion devulve entonces un data frame con los resumenes de cada una de las tablas.

def reporte_calidad_global(dfs, nombres):
    resumen = []
    for df, nombre in zip(dfs, nombres):
        nulos = df.isna().sum().sum()
        duplicados = df.duplicated(keep=False).sum()
        columnas = len(df.columns)
        filas = len(df)

        resumen.append({
            "Dataset": nombre,
            "Filas": filas,
            "Columnas": columnas,
            "Nulos totales": nulos,
            "Duplicados": duplicados,
        })

    reporte = pd.DataFrame(resumen)
    return reporte


print(reporte_calidad_global([ventas, clientes, marketing], ["VENTAS Original", "CLIENTES Original", "MARKETING Original"]))
print(reporte_calidad_global([ventas_clean, clientes_clean, marketing_clean],["VENTAS Limpio", "CLIENTES Limpio", "MARKETING Limpio"]))
````

### 5锔 Transformaci贸n de datos

````python
#Calculo el ingreso de cada uno de los productos (producto *cantidad vendido) y creo nueva columna al df de ingreso

ventas_perf = (
    ventas_clean
    .assign(
        ingreso = ventas_clean["precio"] * ventas_clean["cantidad"]
    )
)



#Agregar m茅tricas por producto
resumen_prod = (
    ventas_perf
    .groupby(
        by="producto",
        dropna=False,
        as_index=False,
        observed=False
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size')
    )
)


print(resumen_prod.head(60))
````

````python
# Calcular percentil 80 de ingreso_total

p80_ingreso = resumen_prod["ingreso_total"].quantile(
    q=0.80,
    interpolation="linear"
)



# Filtrar los productos de alto rendimiento > q80% y ordenarlos de mayor  a menor

ventas_top = (
    resumen_prod
    .query(
        "ingreso_total >= @p80_ingreso",  #filtra filas donde ingreso_total es >= P80
        engine="python"
    )
    .sort_values(
        by=["ingreso_total", "unidades"],
        na_position="last",
        ignore_index=False
    )
)

#muestro los resultados con print
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print(" Productos de ALTO RENDIMIENTO (top 20% por ingreso):")
display(ventas_top.head(60))
````

### 6锔 Agregaci贸n

````python
# Etapa de agregaci贸n, objetos creados hasta el momento
#ventas: dataset original
#ventas_clean: dataset original despues de limpiar datos
#ventas_perf: dataset de ventas_clean con el campo ingreso que es precio*cantidad de cada factura
#resumen_prod: dataset de ventas_perf agrupado por producto y sumado el ingreso de cada venta = ingreso_total
#ventas_top: el detaset de resumen_prod, que tiene productos agrupados, pero solo me quedo con los productos mas vendidos (top del 20%)
#ventas_cat: el dataset de ventas_perf que se copio y se llama ventas_cat
#resumen_cat: del dataset de ventas_perf agrupo por categoria de producto y sumo ingreso_total
ventas_cat = ventas_perf.copy()

resumen_cat = (
    ventas_cat
    .groupby(
        by="categoria",
        dropna=False,
        as_index=False
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        ventas=('ingreso', 'size'),
        precio_promedio=('precio', 'mean')
    )
    .sort_values(
        by='ingreso_total',
        ascending=False,
        na_position='last',
        ignore_index=True
    )
)

# Ticket promedio por venta = ingreso_total / ventas
resumen_cat = resumen_cat.assign(
    ticket_promedio_por_venta = resumen_cat['ingreso_total'] / resumen_cat['ventas']
    # assign: crea/reescribe columnas
)


print("Resumen por categor铆a (ordenado por ingreso_total):")
display(resumen_cat.head(20))
````
